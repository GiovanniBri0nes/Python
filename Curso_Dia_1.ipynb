## Operaciones Matematicas

```python
# 4to Nivel de importancia
# Suma Ordinaria
print(3 + 4)
variable1, variable2 = 20, 40
print(variable1 + variable2)

# Resta Ordinaria
print(2 - 5)
v1 = 10.4
v2 = 15
resultado = v2 - v1
print(resultado)
```

```python
# 3er Nivel dew importancia
# Multiplicacion Oridinaria
print("Multiplicacion Ordinaria")
M1, M2 = 2, 3.14
print(M1 * M2)
# Division Ordinaria
print("-------------------")
print("Divison Ordinaria")
M1, M2, M3, M4, M5 , M6 = 2, 5 , 6, 5, 4, 5
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)
# Division  de Piso
print("-----------------")
print("Division de Piso")
print(M1 // M2)
print(M3 // M4)
print(M5 // M6)

print (6 / 2 * 3)
```

```python
# Potencias
print("Potencias")
print(49 ** (1/2))
P1, P2 = 7, 2
print(P1 ** P2)
```

```python
# Area de un circulo
pi = 3.1416
r = 4
A = pi * (r ** 2)
A
```
