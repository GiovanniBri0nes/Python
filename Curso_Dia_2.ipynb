## Elif

Ahora hagamos un programa en el que preguntamos la calificacion del 0 al 100 y queremos saber su equivalente en el sistema de evaluacion estadounidense. Para simplicidad, basemonos en esto:
- A+ = 100
- A = 90 - 99
- B = 80- 89
- C = 70 - 79
- D = 60 - 69
- F = < 60

```python
cal = 87
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else:
  print("F")
```

## Operadores Logicos con condicionales

```python
# and, or, not
pan = True
queso = True
mayonesa = True
jamon = False
print(pan and queso and mayonesa and jamon)
if not (pan and queso and mayonesa and jamon):
  print("No te puedes hacer el sandwich")
```

```python
cal = -10
if cal >= 0 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else:
    print("F")
else:
  print("Error: La calificacion debe estar entre 0 y 100")
```

```python
# El modulo (%)
# El modulo es un operador matematico que devuelve el residuo de una division
# 10 % 3 da 1, porque 10 dividido entre 3 cabe 3 veces y sobra 1

num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El numero {num} es multiplo de 3 o de 5")

# saber si el numero es par o impar
num2 = 6
if not num % 2 == 0:
  print(f"El numero {num2} es impar")
```

## Ciclo For
Los llamados ciclos for nos ayudan a realizar acciones durante una cantidad determindada de veces

```python
# Imprimir los numeros del 0 al 4
for i in range(5): # Toma 5 como el numero final, pero inicia en 0
  print(i)
```

```python
for num in range(2, 11):
  print(num)
```

```python
# Tambien existe el range(inicio, fin, paso)
for i in range(2, 21, 2):
  print(i)
```

```python
cadena = "Python"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

## While

```python
suma = 0
numeros = int(input("Ingresa el numero que quieres sumar o 0 para salir: "))
while numeros !=0:
  suma += numeros
  numeros = int(input("Ingresa el numero que quieres sumar o 0 para salir: "))
print(f"Tu suma esa: {suma}")
```

```python
# Usar el While como For
numero = 0
while numero <= 12:
  print(numero)
  numero +=1
```

```python
print("Juego: Adivina el Número")
numero_secreto = int(input("Jugador 1, escribe el número secreto (1-20): "))
intentos = 0
max_intentos = 4
while intentos < max_intentos:
  intento = int(input("Adivina el número: "))
  intentos += 1
  if intento == numero_secreto:
    print("¡Correcto!")
    break
  else:
    print("Incorrecto, intenta de nuevo.")
    #Opcional
    if intentos == max_intentos and intento != numero_secreto:
      print("Se acabaron los intentos.")
      print("El número era:", numero_secreto)
```
