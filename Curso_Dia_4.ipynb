## Funciones

## Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"
def despedirse(nombre):
  return f"Adios, {nombre}!"

def mensaje_Completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + "\n" + despedida

print(mensaje_Completo("Giovanni"))
```

```python
# Ahora supongamos que tenemos una heladeria
def pedir_Sabor():
  sabores = ["vainilla", "chocolate", "fresa"]
  print("Sabores dispinibles: ", sabores)
  sabor = input("Elige un sabor: ").strip().lower()

  while sabor not in sabores:
    sabor = input("Elige un sabor (nuevamente): ").strip().lower()
  return sabor

def pedir_Cantidad():
  cantidad = int(input("Cuantas bolas de nieve?: "))
  while cantidad <= 0:
    cantidad = int(input("Cuantas bolas de nieve? (ingresa valor valido): "))
  return cantidad

def pedir_Helado():
  sabor = pedir_Sabor()
  cantidad = pedir_Cantidad()
  print(f"Pedido: {cantidad} bola(s) de {sabor}")
  return sabor, cantidad

total_Pedidos = int(input("Cuantos clientes atenderas?: "))
for i in range(total_Pedidos):
  pedir_Helado()
```

## Funciones Lambda
Una funcion Lambda es una funcion acortada para escribir una funcion de esta forma:

nombre_Funcion = lambda parametros: accion

La diferencia entre las funciones Lambda y las normales, es que suelens er un poco mas limitadas

```python
sumar = lambda x, y : x + y
print(sumar(2, 3))
```

```python
descuento_Lambda = lambda desc, precio : ((precio * (1 - desc)), (precio * desc))
nuevo_Precio = descuento_Lambda(0.4, 500)
print("El precio a pagar es: ", nuevo_Precio[0], "\n")
print("El descuento que se hizo fue de: ", nuevo_Precio[1])
```

```python
from re import A
# Calculadora -> suma, resta, producto, division y potencia
def suma(a, b):
  return a + b

def resta(a, b):
  return a - b

def producto(a, b):
  return a * b

def division(a, b):
  if b == 0:
    return "Error: no se puede dividir entre 0"
  return a / b

def potencia(a, b):
  return a ** b

def menu():
  print("-------Calculadora--------")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. Division")
  print("5. Potencia" )
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("Seleccione una opcion: "))
  except ValueError:
    print("Error: Ingrese un numero valido")
    continue

  if op == 6:
    print("Saliendo...")
    break

  if(op < 1 or op > 6):
    print("Opcion no valida")

  try:
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
  except ValueError:
    print("Ingrese un numero valido")
  if op == 1:
    print("La suma es: ", suma(a, b))
  elif op == 2:
    print("La resta es: ", resta(a, b))
  elif op == 3:
    print("El producto es: ", producto(a, b))
  elif op == 4:
    print("La division es: ", division(a, b))
  elif op == 5:
    print("La potencia es: ", potencia(a, b))
  else:
    print("Opcion no valida")


```
