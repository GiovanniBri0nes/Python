# Existe 1  solo tema que TODO programador independientemente DEBE de saber

JSON es un formato para guardar y transmitir datos. Es muy usado en APIs, archivos de configuracion y almacenamiento de informacion.
(Se parece a un diccionario de Python, pero es solo texto)

Para que sirve?
  - Sirve para convertir objetos de Python -> texto JSON y tambien texto JSON -> objetos de Python
  
  Esto se llama:
    - Serializar: Convertir un objeto de Pyhton en texto JSON
    - Deserializar: Convertir un texto JSON a un objeto de Pyhton

```python
# Funciones principales
# 1) json.dums(obj): Convierte un diccionario u objeto de Python en un string JSON
import json
obj = {
    "nombre" : "Ana",
    "edad": 25}
print(type(obj))
s = json.dumps(obj, ensure_ascii = False)
print(s)
print(type(s))
```

```python
# 2) json.loads(texto_json) : Convierte un string JSON en un diccionario de Python
restaurar = json.loads(s)
print(restaurar)
print(type(restaurar))
```

## Numpy
Provee el tipo ndarray, permite hacer operaciones vectorizadas, broadcasting, utilidades de algebra lineal y funciones matematicas en arrays. Es la base sobre la que se apoyan muchas librerias cientificas en Python

### El arreglo de Nunmpy no es igual a la lista de Python. Este no puede tener diferentes tipos de datos en el mismo

```python
import numpy as np
A = np.array([[1, 2, 3], [4, 5, 6]])
v = np.array([10, 20, 30])
print(A + v)
print("---------------")
prueba = [1, 2 , 3]
prueba_2 = [1, 2, 3]
print(prueba + prueba_2)
```

```python
# Mas funciones de Numpy
z = np.zeros((3, 3))
print(z)
o = np.ones((2, 4))
print(o)
r = np.random.rand(3, 3)
print(r)
```

```python
# Rangos numericos
nums = np.arange(0, 10, 2)
print(nums)
x = np.linspace(0, 1, 5)
print(x)
```

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([6, 7, 8, 9, 10])
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** b)
```

```python
print(a.dot(b)) #productos entre 2 matrices
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[10, 20], [30, 40], [50, 60]])
print(c.dot(d))
```

```python
data = np.array([1, 2, 3, 4, 5])
print(data.max())
print(data.min())
print(data.sum())
print(data.mean())
print(data.std())
```

```python
A = np.array([1, 6, 3, 9, 2])
print(A[A > 5])
```

## Matplotlib

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, marker = "o")
#linestyle, linewidth
plt.title("Titulo de grafica")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```
